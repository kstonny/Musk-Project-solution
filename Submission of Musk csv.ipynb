```python
#import the libraries for splitting the data set into test n train

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=400)

#data is divided into 80:20 ration of tarin:test 
```

```python
#importing bagging tree classifier in which decision tree Classifier is base learner

from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# Instantiating a object of bagging classifier class , BASE ESTIMATOR IS GIVEN AS DECISION TREE.
clf=BaggingClassifier(oob_score=True,n_jobs=-1,n_estimators=20,random_state=400,
                      base_estimator=DecisionTreeClassifier())
```

```python
clf.fit(X_train,y_train) # fitting our bagging tree ensemble using fit method
```

```python
clf.oob_score_  # 97% accuracy on training model
```

```python
clf.score(X_test,y_test) # checking accuracy on testing data.

#97 % accuracy model is good to use 
```

```python
#let's find the best n_estimator 
for w in range(10,300,20):
    clf=BaggingClassifier(oob_score=True,n_jobs=-1,n_estimators=w,random_state=400,
                          base_estimator=DecisionTreeClassifier())
    clf.fit(X_train,y_train)
    oob=clf.oob_score_ 
    print ('For n_estimators = '+str(w))
    print ('OOB score is '+str(oob))
    print ('************************')
```

```python
# Let's run the classifier again with higherst oob score's tree now. 
#Finalizing on a tree model with 250 trees as the accuracy is highest in it.
clf=BaggingClassifier(oob_score=True,n_jobs=-1,n_estimators=250,random_state=400,
                      base_estimator=DecisionTreeClassifier())
clf.fit(X_train,y_train) # fitting bagged tree classifier  model 
```

```python
#As we have to make the classificaton on data points its better to use Multi Layer Preceptrons / ANN.

# we can juse the significant feature of pre trained model lets say 15 feature we select.
```

```python
# independent variable(X) is created for ANN model where y is the classes only i.e.

X = data[["f36","f124","f9","f132","f151","f126","f122","f140","f66","f50","f63","f33","f43","f163","f116"]]
```

```python
# Splitting the dataset into the Training set and Test set i.e. 80:20 ratio.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling for making the data point scaled between -1 to 1
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Now let's make the ANN!

# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initialising the ANN
classifier = Sequential()
```

```python
# Adding the input layer and the first hidden layer i.e 15 significant variables and 8 neural structures 
classifier.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'relu', input_dim = 15))
```

```python
# Adding the second hidden layer of 6 nureal structure
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
```

```python
# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set and validation
history = classifier.fit(X_train, y_train, batch_size = 30, validation_split=0.2, epochs = 50, verbose=1)
```

#Classifier is the h5 model that we have trained.

# Graphical Representation

```python
print(history.history.keys())
```

```python
#  "Accuracy" graph both for training and validation.
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
# "Loss" graph for both training and validation.
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

# Final Performance Measures

```python
# classifier is the h5 model that we have traind now
# Making predictions and evaluating the model
# Predicting the Test set results

y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
#assigning the True Positive , True Negative , False positive and false negative
TP = cm[1,1]
TN = cm[0,0]
FP = cm[0,1]
FN = cm[1,0]
```

```python
# Validation Accuracy of the model.
Accuracy = print("Validation Accuracy is" , ((TP+TN)/float(TP+TN+FP+FN)*100))
```

```python
#error rate 

error_rate = print("Loss is", (FP+FN)/float(TP+TN+FP+FN)*100)
```

```python
#Precition
precision = TP / float(TP + FP)

print("Precition is" ,precision)
```

```python
#Recall 

recall = TP / float(TP + FN)

print("Recall/Senstivity is ", recall)
```

```python
#F1 Score 

f1score = 2*((precision*recall)/float((precision+recall)))

print("f1score is " , f1score)
```
